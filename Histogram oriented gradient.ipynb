```python
import cv2 as c
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from skimage.feature import hog 
from skimage import data ,exposure
```

```python
img=c.imread("b.jpg")
c.imshow("IMAGE",img)
c.waitKey(0)
```

# HOG

```python
fd,hog_image=hog(img,orientations=9,pixels_per_cell=(6,6),cells_per_block=(5,5),visualize=True,channel_axis=2)
```

# RESCALE

```python
cale=exposure.rescale_intensity(hog_image,in_range=(55,70))
```

# plot

```python
figure,(a1,a2)=plt.subplots(1,2,figsize=(12,6),sharex=True,sharey=True)
a1.axis('off')
a1.imshow(img)
a1.set_title('input image')
a2.axis('off')
a2.imshow(cale)
a2.set_title("hog image")
plt.show()
```

```python
hog=c.HOGDescriptor()
hog.setSVMDetector(c.HOGDescriptor_getDefaultPeopleDetector())
image=c.imread("peoples.jpg")
image=imutils.resize(image,width=min(400,image.shape[1]))
(regions,img)=hog.detectMultiScale(image,winStride=(2,2),padding=(4,4),scale=1.05)
for(x,y,w,h) in regions:
    c.rectangle(image,(x,y),(x+w,y+h),(0,0,255),10)
c.imshow("po",image)
c.waitKey(0)
```
